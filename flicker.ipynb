```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import urllib.request
from urllib.request import Request,urlopen
import requests
import os
import time

#this is for what u want to search......
name=input("enter the name what u want to search = ")

#this is for open page in incognito mode .......
chrome_options=Options()
chrome_options.add_argument('-incognito')
driver=webdriver.Chrome(executable_path="C:\\Users\\User\\Desktop\\DRIVERS\\chromedriver.exe",options=chrome_options)
driver.implicitly_wait(10)
url="https://www.flickr.com/search/?text=" + name

#this is for create folder name with name what u search.......
folderName=url.split('=')[-1]

driver.get(url)
driver.maximize_window()
p=driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var p=document.body.scrollHeight;return p;")
posts=[]
match=False
while(match==False):
    page_count=p
    time.sleep(4)
    p=driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var p=document.body.scrollHeight;return p;")
    if page_count==p:
        match=True
    try:
        driver.find_element_by_class_name("alt").click()

    except:
        pass
    links=driver.find_elements_by_class_name('overlay')
    for i in links:
        try:
            link=i.get_attribute('href')
        except:
            pass
        if link != None and '#comments' not in link:
            if link not in posts:
                posts.append(link)
img_links=[]
for post in posts:
        driver.get(post)
        try:
            p=driver.find_elements_by_tag_name('img')[1]
            m=p.get_attribute("src")
            img_links.append(m)
        except:
            pass
for link in img_links:
    folder="C:\\Users\\User\\Desktop\\flicker\\" + folderName
    os.makedirs(folder, exist_ok = True)
    data = requests.get(link)    
    file = open(os.path.join(folder, link.split("/")[-2])+'.jpg', "wb")
    file.write(data.content)
```
